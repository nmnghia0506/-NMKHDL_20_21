**Lợi ích khi trả lời được câu hỏi:** Khách hàng thì có nhiều đối tượng với nhu cầu mỗi người mỗi khác nhau, có người tửu lượng mạnh thì thích các loại rượu có nồng độ cao, có người thì thích những loại có nồng độ yếu hơn. Mục tiêu ở đây là giúp người mua có thể biết được các thương hiệu có giá thành ổn nhất đối với từng mức nồng độ, từ đó đưa ra các lựa chọn phù hợp với túi tiền của mình.

<h3 style='color:purple'> Các bước thực hiện</h3>

Chỉ lấy ra các cột cần thiết

```python
filter_df = df.loc[:,['Ma_sp','Ten_sp','Nong_do','Thuong_hieu','Gia']]
filter_df.head(5)
```

Ta sẽ tiến hành chia nồng độ rượu vang thành 4 mức độ theo tiêu chuẩn  
Tham khảo: [Wine's Alcohol levels](https://www.wineinvestment.com/learn/magazine/2019/06/wines-alcohol-levels-explained/)

| Phân loại | Nồng độ |
| --- | --- |
| **Nồng độ thấp** | < 12.5% |
| **Nồng độ vừa phải** | 12.5 - 13.5% |
| **Nồng độ cao** | 13.5 - 14.5% |
| **Nồng độ rất cao** | > 14.5% |

```python
def alcohol_classify(degree):
  if degree < 12.5:
    return 'Nồng độ thấp'
  elif 12.5 <= degree <= 13.5:
    return 'Nồng độ vừa phải'
  elif 13.5 <= degree <= 14.5:
    return 'Nồng độ cao'
  return 'Nồng độ rất cao'

filter_df['alcohol_classify'] = filter_df['Nong_do'].apply(alcohol_classify)
filter_df
```

Phân bố số rượu theo các mức nồng độ theo phần trăm tổng số rượu

```python
filter_df['alcohol_classify'].value_counts() / len(filter_df) * 100
```

Rượu có mức nồng độ vừa phải chiếm số lượng nhiều nhất với hơn 50% tổng số rượu

Tính giá rượu trung bình theo từng mức nồng độ và thương hiệu 

```python
price_by_ac_brand_df = filter_df.groupby(['alcohol_classify', 'Thuong_hieu'], as_index = False)['Gia'].mean()
price_by_ac_brand_df
```

Lấy ra 5 thương hiệu có giá trung bình thấp nhất theo từng nồng độ

```python
n_cheapest_df = price_by_ac_brand_df.groupby('alcohol_classify')['Gia'].nsmallest(5).reset_index(level = 1)
original_index = n_cheapest_df.pop('level_1')
n_cheapest_df = price_by_ac_brand_df.loc[original_index].reset_index(drop = True)
n_cheapest_df
```

Trực quan lên biểu đồ 5 thương hiệu có giá rượu trung bình thấp nhất theo từng mức nồng độ.

```python
alcohol_classify_order = ['Nồng độ thấp', 'Nồng độ vừa phải', 'Nồng độ cao', 'Nồng độ rất cao']
fig,axes = plt.subplots(4,1, figsize = (16,10), sharex = True)
for i,ax in enumerate(axes.flat):
  sns.barplot(n_cheapest_df.loc[n_cheapest_df['alcohol_classify'] == alcohol_classify_order[i]], x = 'Gia', y = 'Thuong_hieu', ax = ax)
  if i == len(axes) - 1:
    ax.set_xlabel('Giá')
  else:
    ax.set_xlabel('')
  ax.set_ylabel('Thương hiệu')
  ax.set_title(alcohol_classify_order[i])
plt.tight_layout()
```

Ứng với 4 mức nồng độ, ta đã xác định được 5 thương hiệu có giá thành trung bình thấp nhất. Dựa vào biểu đồ bên trên ta rút ra được các nhận xét:
* Có vẻ là mức nồng độ tăng dần đồng nghĩa với việc giá rượu cũng tăng theo, với trường hợp ngoại lệ là mức nồng độ vừa phải có giá trung bình thấp nhất trong tất cả các mức.
* Thương hiệu có giá rượu trung bình thấp nhất ở mức nồng độ vừa phải và mức nồng độ cao đều thuộc về Vina Marquez, phù hợp với đối tượng khách hàng chuộng các sản phẩm có nồng độ không quá thấp cũng như không quá cao.
* Cantine Sgarzi Luigi là thương hiệu có giá thành rẻ nhất ở mức nồng độ thấp và cũng xếp thứ tư ở thương hiệu có giá thấp nhất ở mức nồng độ rất cao.

### <h2 style='color:#8edc55'>Câu 3: Thống kê số lương sản phẩm mà mỗi loại nho xuất hiện. Loại nho nào xuất hiện trong nhiều loại rượu nhất?  </h2>

**Lợi ích khi trả lời câu hỏi**
- Đối với một người mua rượu, biết được loại nho khác nhau dùng làm rượu và loại nho nào được dùng làm rượu nhiều nhất, bởi vì chất lượng rượu phụ thuộc phần lớn vào giống nho.
- Người bán khảo sát thị trường để trồng loại nho phù hợp với sở thích người mua

```python
def is_nan(x):
    return (x != x)
```

<h3 style='color:#c177e4'> Bước 1: Xử lý </h3>

Cột **Giong_nho** có dạng multivalue được ngăn cách bời dấu ',', do đó ta cần lấy hết các giá trị đơn và lưu vào một danh sách. Sau đó đếm các giá trị đó và lưu vào một dataframe

```python
temp_df = df[['Giong_nho']]
items = ['Giong_nho']
temp_df = temp_df.dropna()

# itemLists chứa giá trị đơn của cột Giong_nho
itemLists = []

# label danh sách index để tạo df
label = []
for i in range(0,len(items)):
    s = temp_df[items[i]].tolist()
    s_split = []
    for i in range(0,len(s)):
        if is_nan(s[i]) == 0:
            s_split.append(re.split(',',s[i]))
    itemLists.append(s_split)
    
    tempList = pd.Series(s_split).explode().tolist()
    for j in range(0,len(tempList)):
        label.append(tempList[j])
itemLists =itemLists[0]

label = [*set(label)]
for i in range (0,len(label)):
    label[i] = label[i].strip()
#     l[i] = re.split('và',l[i])
    label[i] = re.split('[–-]',label[i])
tempList = []
for i in range (0,len(label)):
    for j in range (0,len(label[i])):
        if label[i][j] != 'Ấn tượng đặc biệt có trong chai Rượu Vang Domaine De Savagny Cotes Du Jura Vin Jaune bởi hương vị đặc biệt mang một phong cách mới mẻ của nhà sản xuất Domaine de Savagny. Rượu phần nào thể hiện được các tính con người bạn qua quá trình thưởng thức.':
            tempList.append(label[i][j])
            tempList[len(tempList) - 1] = tempList[len(tempList) - 1].strip()
            tempList[len(tempList) - 1] = re.findall('[a-z A-Z]+',tempList[len(tempList) - 1] )
label = tempList
tempList = []
for i in range (0,len(label)):
    for j in range (0,len(label[i])):
            if len(label[i][j]) >= 3:
                tempList.append(label[i][j])
                tempList[len(tempList) - 1] = tempList[len(tempList) - 1].strip()
label = tempList 
label = [*set(label)]

# num_val_list chứa số lượng của các giống nho
num_vals_list = []
for i in range(0,len(label)):
    num_vals_list.append(0)
tempList = []
for i in range (0,len(itemLists)):
    for j in range (0,len(itemLists[i])):
            tempList.append(itemLists[i][j])
            tempList[len(tempList) - 1] = tempList[len(tempList) - 1].strip()
itemLists = tempList

for i in range(0,len(itemLists)):
    for j in range(0,len(label)):
        if itemLists[i] == label[j]:
            num_vals_list[j] += 1

popular = pd.Series(data = num_vals_list, index = label)
popular = popular.sort_values(ascending=False)
popular = popular.to_frame()
popular.columns = ["So_luong"]
```

<h2 style='color:#8edc55'> Câu 5: Có mối liên hệ nào giữa vùng làm vang và giá rượu không? Thống kê những vùng có mức giá trung bình thấp nhất và cao nhất.</h2>

**Lợi ích khi trả lời được câu hỏi:** Mỗi vùng làm vang sẽ sản xuất ra những loại rượu có chất lượng khác nhau. Nên nếu thống kê được mức giá trung bình của từng vùng sẽ giúp người mua dễ dàng biết được mức giá của sản phẩm đến từ những vùng đó có hợp lý không, và dễ dàng chọn lựa hơn.

<h3 style='color:#c177e4'> 1. Chọn dữ liệu cần thiết</h3>

```python
#Chọn ra 2 cột cần thiết là Vung_lam_vang và Gia
region_df = df[['Vung_lam_vang', 'Gia']]
region_df
```

```python
#Thống kê số lượng sản phẩm theo vùng làm vang
region = region_df['Vung_lam_vang'].value_counts()
region
```

Ta có thể thấy có tổng cộng 351 vùng làm vang trong 3934 sản phẩm. Nhưng nhiều vùng chỉ có duy nhất 01 sản phẩm hoặc rất ít sản phẩm, nên ta sẽ cho tất cả những vùng có số sản phẩm nhỏ hơn 30 vào một vùng `Others`.

```python
#Chọn những vùng có số sản phẩm ít hơn 30
non_popular_region = region[region < 30]
non_popular_region
```

```python
#Gộp những vùng có số sản phẩm ít hơn 30 vào vùng 'Others'
region_df['Vung_lam_vang'] = region_df['Vung_lam_vang'].apply(lambda x: 'Others' if x in non_popular_region else x)
region_df
```

<h3 style='color:#c177e4'>2.Trực quan hoá và đánh giá kết quả</h3>

```python
plt.figure(figsize=(18,10))
sns.barplot(data=region_df, y='Gia', x = 'Vung_lam_vang', estimator=np.mean, order=region_df.groupby(['Vung_lam_vang'])['Gia'].mean().sort_values(ascending=False).index)
plt.title('Phân bố giá rượu trung bình của một số vùng làm vang', fontsize=14)
plt.xticks(rotation=45)
plt.xlabel('Vùng làm vang', fontsize=14)
plt.ylabel('Giá rượu (VNĐ)', fontsize=14)
plt.grid(True)
plt.show()
```

**Kết luận:**
* Có thể thấy rằng mức vùng làm vang có ảnh hưởng tới mức giá trung bình của sản phẩm.
* 6 vùng làm vang có giá trung bình đắt nhất là Bourgogne (Pháp), Piedmont (Ý), Napa Valley (Hoa Kỳ), Rhone Valley (Pháp), Barossa Valley (Úc) và Bordeaux (Pháp). Đây đều là những vùng làm rượu vang lâu đời và nổi tiếng thế giới.
* Các vùng rẻ nhất là Maule Valley (Chile), Languedoc (Pháp), Sicilia (Ý), Casablanca Valley (Chile) và Central Valley (Chile).
* Các vùng vang khác có giá không chênh lệch nhau quá nhiều.
